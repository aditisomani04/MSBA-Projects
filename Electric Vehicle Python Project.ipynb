## How many models of Electric Vehicle Type are present in the data frame ? 

```python
EV_Type = df[['Electric Vehicle Type','Model']].groupby('Electric Vehicle Type').count().reset_index()
EV_Type
```

### We can see that there are only 2 types of Electric Vehicle models. And the majority is Battery Electric Vehicle model. The less famous type is Plug-in Hybrid EV.

### Let us see the pictorial representation of above output in the form of a circular chart

```python
explode = [0,0.1]
plt.pie(list(EV_Type['Model']),colors = ['#90EE90','#023020'], 
        labels= list(EV_Type['Electric Vehicle Type']),
        explode=explode, autopct='%.2f%%')
```

### From the above pie chart, we can understand that there are 75.65% Battery EV and 24.35% Plug-in Hybrid EV.

###  

## How many models were released over a period of 25 years from 1997-2023 ?

```python
EV_Rise = df[['Model Year','Model']].groupby('Model Year').count().reset_index()
EV_Rise
```

### Let us see the bar graphical representation of Electric Vehicles over a period of 25 years from the year 1997-2023

```python
sns.set(rc = {'figure.figsize':(10,5)})
sns.barplot(data= EV_Rise,x='Model Year',y='Model')
plt.xlabel("Release_Year")
plt.ylabel("Number of Models_Released")
plt.title("Electric Vehicle Models Released (1997-2023)")
```

### From the above bar graph, we can see the rise of Electric Vehicles and it will continue to increase in the near future as the natural resource petrol and diesel get exhausted. We must move towards green and pollution free options of commuting, also considering the global warming issue in mind.
